```python
# Efficient linear algebra algorithms and data structures 
import numpy as np

# Reading CSV data
import csv

# For creating graphs/plots
import matplotlib.pyplot as plt

# For creating tables and manipulating data
import pandas as pd

# For actual machine learning
from sklearn import linear_model, datasets
```

```python
# Get the training data
rows = []
with open('/home/local/ANT/naddeo/workspace/titanic-machine-learning/data/train.csv', 'rb') as trainingData:
    for row in csv.reader(trainingData):
        rows.append(row)
        
header = rows[0]
data = np.array(rows[1:])
        

pd.DataFrame(data, columns=header)
```

```python
# Take a subset of the data and call it X
# For starting, let's try something that is already numeric and probably coorelated with survival
# gender and price at first
X =  data[:,  [4,9]]

# Men will be 1 and women will be 2
X[X == 'male'] = 1
X[X == 'female'] = 2
X = X.astype(np.float)

X
```

```python
# Whether or not they survived
y=data[:, 1].astype(np.float)
y
```

```python
# Split the data set into two data sets of people who survived and people who did not

fullData =  np.column_stack((X, y)) 

# Get the indicies of everyone who lived/died
survivedIndexes = np.squeeze(fullData[:,2])==1
deadIndexes = np.squeeze(fullData[:,2])==0

# Use those indicies to split into two data sets that we can plot separately
survivors = fullData[survivedIndexes, :]
deaders = fullData[deadIndexes, :]
```

```python
# See if there really is a correlation here
# It would appear you were probably going to survive if you were a wealthy woman
fig = plt.figure()
ax1 = fig.add_subplot(111)

plt.ylabel('ticket price')
plt.xlabel('gender') 



plt.scatter(survivors[:, 0], survivors[:, 1], marker='s', label='survived')
plt.scatter(deaders[:, 0], deaders[:, 1], label='dead')
plt.legend(loc='upper center');
plt.show()
```

```python
# Do some ML
model = linear_model.LogisticRegression()
model.fit(X, y)
```

```python
# Make predictions

poorGuy = np.array( [[1, 10]])
model.predict(poorGuy)
```

```python
# Rich man. Needs to buy a ticket for at least $154 to survive it seems
richGuy = np.array( [[1, 154]])
model.predict(richGuy)
```

```python
# Rich woman

richWoman = [[2, 0]]
model.predict(richWoman)
```
